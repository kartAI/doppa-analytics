```python
from src.persistence import create_duckdb_context
```

```python
db = create_duckdb_context()
df = db.execute(
    f"SELECT * FROM read_parquet('az://benchmarks/query_id=duckdb-bbox-filtering/run_id=2026-02-23-FILDPV/iteration=*/*.parquet')").fetchdf()
df.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Normalize column naming and add an iteration column if needed.
if "iteration" not in df.columns:
    for candidate in ("iteration_id", "iter", "iteration_idx"):
        if candidate in df.columns:
            df = df.rename(columns={candidate: "iteration"})
            break

if "iteration" not in df.columns:
    raise KeyError(
        "Missing `iteration` column. Add it to the query or rename the column accordingly."
    )

required_cols = {"wall_clock_time", "cpu_percentage", "iteration"}
missing = required_cols - set(df.columns)
if missing:
    raise KeyError(f"Missing required columns: {sorted(missing)}")

plot_df = df.copy()
plot_df["wall_clock_time"] = pd.to_numeric(plot_df["wall_clock_time"], errors="coerce")
plot_df["cpu_percentage"] = pd.to_numeric(plot_df["cpu_percentage"], errors="coerce")
plot_df = plot_df.dropna(subset=["wall_clock_time", "cpu_percentage", "iteration"])

plt.figure(figsize=(10, 6))
for iteration, group in plot_df.sort_values("wall_clock_time").groupby("iteration"):
    plt.plot(group["wall_clock_time"], group["cpu_percentage"], label=str(iteration))


plt.title("CPU Percentage vs Wall Clock Time (per iteration)")
plt.xlabel("Wall Clock Time")
plt.ylabel("CPU Percentage")
plt.legend(title="Iteration", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
