```python
import matplotlib.pyplot as plt

from src.persistence import create_duckdb_context
```

```python
db = create_duckdb_context()
```

```python
df = db.execute(f"SELECT * FROM read_parquet('az://benchmarks/query_id=main/run_id=2026-02-19-77JSP2/iteration=*/*.parquet')").fetchdf()
```

```python
df.head()
```

```python
# Get all unique iterations
iterations = sorted(df['iteration'].unique())
print(f"Plotting all {len(iterations)} iterations: {iterations}")
```

```python
# Plot CPU Percentage for all iterations on the same plot
plt.figure(figsize=(14, 7))
for iteration in iterations:
    iter_df = df[df['iteration'] == iteration]
    plt.plot(iter_df['wall_clock_time'], iter_df['cpu_percentage'], marker='o', linewidth=2, markersize=4, alpha=0.7, label=f'Iteration {iteration}')

plt.xlabel('Wall Clock Time', fontsize=12)
plt.ylabel('CPU Percentage (%)', fontsize=12)
plt.title('CPU Percentage vs Wall Clock Time - All Iterations', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Plot RSS for all iterations on the same plot
plt.figure(figsize=(14, 7))
for iteration in iterations:
    iter_df = df[df['iteration'] == iteration]
    plt.plot(iter_df['wall_clock_time'], iter_df['rss'], marker='o', linewidth=2, markersize=4, alpha=0.7, label=f'Iteration {iteration}')

plt.xlabel('Wall Clock Time', fontsize=12)
plt.ylabel('RSS (Resident Set Size)', fontsize=12)
plt.title('RSS vs Wall Clock Time - All Iterations', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Calculate and print std for each iteration
print("Standard Deviations by Iteration:")
print("=" * 60)
for iteration in iterations:
    iter_df = df[df['iteration'] == iteration]
    cpu_std = iter_df['cpu_percentage'].std()
    rss_std = iter_df['rss'].std()
    print(f"Iteration {iteration}: CPU σ={cpu_std:.4f}, RSS σ={rss_std:.4f}")
```

```python
# Create a scatter plot: Normalized CPU Percentage vs Normalized RSS
import numpy as np

fig, ax = plt.subplots(figsize=(14, 8))

# Create a color map for iterations - dynamically scale for any number of iterations
if len(iterations) <= 10:
    colors = plt.cm.tab10(np.linspace(0, 1, len(iterations)))
else:
    colors = plt.cm.tab20(np.linspace(0, 1, min(len(iterations), 20)))
    if len(iterations) > 20:
        colors = plt.cm.hsv(np.linspace(0, 1, len(iterations)))

for idx, iteration in enumerate(iterations):
    iter_df = df[df['iteration'] == iteration].reset_index(drop=True)

    # Normalize both metrics to 0-100 scale
    cpu_normalized = (iter_df['cpu_percentage'] - iter_df['cpu_percentage'].min()) / (iter_df['cpu_percentage'].max() - iter_df['cpu_percentage'].min()) * 100
    rss_normalized = (iter_df['rss'] - iter_df['rss'].min()) / (iter_df['rss'].max() - iter_df['rss'].min()) * 100

    ax.scatter(cpu_normalized, rss_normalized, s=100, alpha=0.6, label=f'Iteration {iteration}', color=colors[idx], edgecolors='black', linewidth=0.5)

ax.set_xlabel('Normalized CPU Percentage (0-100)', fontsize=12)
ax.set_ylabel('Normalized RSS (0-100)', fontsize=12)
ax.set_title('Distribution: Normalized CPU Percentage vs Normalized RSS - All Iterations', fontsize=14, fontweight='bold')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Box plot comparison - CPU Percentage across iterations
import pandas as pd

cpu_data = []
labels = []
for iteration in iterations:
    iter_df = df[df['iteration'] == iteration]
    cpu_data.append(iter_df['cpu_percentage'].values)
    labels.append(f'Iter {iteration}')

fig, ax = plt.subplots(figsize=(14, 7))
bp = ax.boxplot(cpu_data, labels=labels, patch_artist=True)

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('CPU Percentage (%)', fontsize=12)
ax.set_title('CPU Percentage Distribution by Iteration - Box Plot', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Box plot comparison - RSS across iterations
rss_data = []
for iteration in iterations:
    iter_df = df[df['iteration'] == iteration]
    rss_data.append(iter_df['rss'].values)

fig, ax = plt.subplots(figsize=(14, 7))
bp = ax.boxplot(rss_data, labels=labels, patch_artist=True)

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('RSS (Resident Set Size)', fontsize=12)
ax.set_title('RSS Distribution by Iteration - Box Plot', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Violin plot - CPU Percentage across iterations
fig, ax = plt.subplots(figsize=(14, 7))
parts = ax.violinplot(cpu_data, positions=range(len(iterations)), widths=0.7, showmeans=True, showmedians=True)

for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(colors[i])
    pc.set_alpha(0.7)

ax.set_xticks(range(len(iterations)))
ax.set_xticklabels(labels, rotation=45)
ax.set_ylabel('CPU Percentage (%)', fontsize=12)
ax.set_title('CPU Percentage Distribution by Iteration - Violin Plot', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Violin plot - RSS across iterations
fig, ax = plt.subplots(figsize=(14, 7))
parts = ax.violinplot(rss_data, positions=range(len(iterations)), widths=0.7, showmeans=True, showmedians=True)

for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(colors[i])
    pc.set_alpha(0.7)

ax.set_xticks(range(len(iterations)))
ax.set_xticklabels(labels, rotation=45)
ax.set_ylabel('RSS (Resident Set Size)', fontsize=12)
ax.set_title('RSS Distribution by Iteration - Violin Plot', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Density/KDE plot - CPU Percentage across iterations
fig, ax = plt.subplots(figsize=(14, 7))

for idx, iteration in enumerate(iterations):
    iter_df = df[df['iteration'] == iteration]
    iter_df['cpu_percentage'].plot.kde(ax=ax, label=f'Iteration {iteration}', linewidth=2, color=colors[idx])

ax.set_xlabel('CPU Percentage (%)', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('CPU Percentage Distribution Density - All Iterations', fontsize=14, fontweight='bold')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Density/KDE plot - RSS across iterations
fig, ax = plt.subplots(figsize=(14, 7))

for idx, iteration in enumerate(iterations):
    iter_df = df[df['iteration'] == iteration]
    iter_df['rss'].plot.kde(ax=ax, label=f'Iteration {iteration}', linewidth=2, color=colors[idx])

ax.set_xlabel('RSS (Resident Set Size)', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('RSS Distribution Density - All Iterations', fontsize=14, fontweight='bold')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Summary statistics table
summary_stats = []
max_cpu_list = []
max_rss_list = []

for iteration in iterations:
    iter_df = df[df['iteration'] == iteration]
    max_cpu = iter_df['cpu_percentage'].max()
    max_rss = iter_df['rss'].max()

    max_cpu_list.append(max_cpu)
    max_rss_list.append(max_rss)

    summary_stats.append({
        'Iteration': iteration,
        'CPU Mean': f"{iter_df['cpu_percentage'].mean():.2f}",
        'CPU Std': f"{iter_df['cpu_percentage'].std():.2f}",
        'CPU Max': f"{max_cpu:.2f}",
        'RSS Mean': f"{iter_df['rss'].mean():.2f}",
        'RSS Std': f"{iter_df['rss'].std():.2f}",
        'RSS Max': f"{max_rss:.2f}",
    })

summary_df = pd.DataFrame(summary_stats)
print("\nSummary Statistics by Iteration:")
print("=" * 100)
print(summary_df.to_string(index=False))
```

```python
# Plot Maximum CPU Percentage across iterations
fig, ax = plt.subplots(figsize=(14, 7))

bars = ax.bar(range(len(iterations)), max_cpu_list, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
ax.set_xticks(range(len(iterations)))
ax.set_xticklabels(labels, rotation=45)
ax.set_ylabel('Maximum CPU Percentage (%)', fontsize=12)
ax.set_title('Maximum CPU Percentage by Iteration', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

# Add value labels on top of bars
for i, (bar, val) in enumerate(zip(bars, max_cpu_list)):
    ax.text(bar.get_x() + bar.get_width()/2, val + 1, f'{val:.2f}%',
            ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Plot Maximum RSS across iterations
fig, ax = plt.subplots(figsize=(14, 7))

bars = ax.bar(range(len(iterations)), max_rss_list, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
ax.set_xticks(range(len(iterations)))
ax.set_xticklabels(labels, rotation=45)
ax.set_ylabel('Maximum RSS (Resident Set Size)', fontsize=12)
ax.set_title('Maximum RSS by Iteration', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

# Add value labels on top of bars
for i, (bar, val) in enumerate(zip(bars, max_rss_list)):
    ax.text(bar.get_x() + bar.get_width()/2, val + max(max_rss_list)*0.01, f'{val:.0f}',
            ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Comparison plot: Both Max CPU and Max RSS (normalized) on same plot
fig, ax = plt.subplots(figsize=(14, 7))

# Normalize RSS to 0-100 for fair comparison
rss_normalized_max = [(val - min(max_rss_list)) / (max(max_rss_list) - min(max_rss_list)) * 100 for val in max_rss_list]

x = np.arange(len(iterations))
width = 0.35

bars1 = ax.bar(x - width/2, max_cpu_list, width, label='Max CPU (%)', color='#1f77b4', alpha=0.8, edgecolor='black', linewidth=1)
bars2 = ax.bar(x + width/2, rss_normalized_max, width, label='Max RSS (Normalized 0-100)', color='#ff7f0e', alpha=0.8, edgecolor='black', linewidth=1)

ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.set_ylabel('Value', fontsize=12)
ax.set_title('Maximum CPU Percentage vs Maximum RSS (Normalized) - Comparison', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# Line plot: Peak values trend across iterations
fig, ax = plt.subplots(figsize=(14, 7))

ax.plot(range(len(iterations)), max_cpu_list, marker='o', linewidth=2.5, markersize=8,
        label='Max CPU Percentage', color='#1f77b4', alpha=0.8)
ax.plot(range(len(iterations)), rss_normalized_max, marker='s', linewidth=2.5, markersize=8,
        label='Max RSS (Normalized)', color='#ff7f0e', alpha=0.8)

ax.set_xticks(range(len(iterations)))
ax.set_xticklabels(labels, rotation=45)
ax.set_ylabel('Value', fontsize=12)
ax.set_title('Peak Performance Metrics Trend - Similarity Analysis', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Calculate similarity metrics
cpu_max_std = np.std(max_cpu_list)
cpu_max_mean = np.mean(max_cpu_list)
cpu_max_cv = (cpu_max_std / cpu_max_mean) * 100  # Coefficient of variation

rss_max_std = np.std(max_rss_list)
rss_max_mean = np.mean(max_rss_list)
rss_max_cv = (rss_max_std / rss_max_mean) * 100  # Coefficient of variation

print("\n" + "="*80)
print("SIMILARITY ANALYSIS - Peak Performance Metrics")
print("="*80)
print(f"\nMaximum CPU Percentage:")
print(f"  Mean:                {cpu_max_mean:.2f}%")
print(f"  Std Dev:             {cpu_max_std:.2f}%")
print(f"  Coefficient of Var:  {cpu_max_cv:.2f}%")
print(f"  Min:                 {min(max_cpu_list):.2f}%")
print(f"  Max:                 {max(max_cpu_list):.2f}%")
print(f"  Range:               {max(max_cpu_list) - min(max_cpu_list):.2f}%")

print(f"\nMaximum RSS:")
print(f"  Mean:                {rss_max_mean:.2f}")
print(f"  Std Dev:             {rss_max_std:.2f}")
print(f"  Coefficient of Var:  {rss_max_cv:.2f}%")
print(f"  Min:                 {min(max_rss_list):.2f}")
print(f"  Max:                 {max(max_rss_list):.2f}")
print(f"  Range:               {max(max_rss_list) - min(max_rss_list):.2f}")

print(f"\n{'INTERPRETATION:'}")
print(f"  CV < 10%:  Very similar runs (high consistency)")
print(f"  CV 10-20%: Similar runs (good consistency)")
print(f"  CV > 20%:  Variable runs (lower consistency)")

if cpu_max_cv < 10:
    print(f"\n✓ CPU Performance: VERY SIMILAR across iterations (CV={cpu_max_cv:.2f}%)")
elif cpu_max_cv < 20:
    print(f"\n✓ CPU Performance: SIMILAR across iterations (CV={cpu_max_cv:.2f}%)")
else:
    print(f"\n✗ CPU Performance: VARIABLE across iterations (CV={cpu_max_cv:.2f}%)")

if rss_max_cv < 10:
    print(f"✓ Memory Performance: VERY SIMILAR across iterations (CV={rss_max_cv:.2f}%)")
elif rss_max_cv < 20:
    print(f"✓ Memory Performance: SIMILAR across iterations (CV={rss_max_cv:.2f}%)")
else:
    print(f"✗ Memory Performance: VARIABLE across iterations (CV={rss_max_cv:.2f}%)")
print("="*80)
```
