```python
import matplotlib.pyplot as plt

from src.persistence import create_duckdb_context
```

```python
db = create_duckdb_context()
```

```python
df = db.execute(f"SELECT * FROM read_parquet('az://benchmarks/main-02bd6d90-3d34-48ef-87f7-1a74a4acab9c-1.parquet')").fetchdf()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['wall_clock_time'], df['cpu_percentage'], marker='o', linewidth=2, markersize=4, alpha=0.7)
plt.xlabel('Wall Clock Time')
plt.ylabel('CPU Percentage')
plt.title('CPU Percentage vs Wall Clock Time')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['wall_clock_time'], df['rss'], marker='o', linewidth=2, markersize=4, alpha=0.7)
plt.xlabel('Wall Clock Time')
plt.ylabel('RSS (Resident Set Size)')
plt.title('RSS vs Wall Clock Time')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
cpu_std = df['cpu_percentage'].std()
rss_std = df['rss'].std()

print(f"Standard Deviation of CPU Percentage: {cpu_std}")
print(f"Standard Deviation of RSS: {rss_std}")
```

```python
cpu_mean = df['cpu_percentage'].mean()
rss_mean = df['rss'].mean()

# Normalize RSS to a 0-100 scale for comparison with CPU percentage
rss_normalized = (df['rss'] - df['rss'].min()) / (df['rss'].max() - df['rss'].min()) * 100
rss_normalized_mean = rss_normalized.mean()
rss_normalized_std = rss_normalized.std()

fig, ax = plt.subplots(figsize=(14, 7))

# Plot CPU Percentage with std bands
ax.plot(df['wall_clock_time'], df['cpu_percentage'], marker='o', linewidth=2, markersize=4,
        alpha=0.8, label='CPU Percentage', color='#1f77b4')
ax.fill_between(df['wall_clock_time'],
                 df['cpu_percentage'] - cpu_std,
                 df['cpu_percentage'] + cpu_std,
                 alpha=0.2, color='#1f77b4', label=f'CPU ±1 Std Dev (σ={cpu_std:.2f})')

# Plot normalized RSS with std bands
ax2 = ax.twinx()
ax2.plot(df['wall_clock_time'], rss_normalized, marker='s', linewidth=2, markersize=4,
         alpha=0.8, label='RSS (Normalized)', color='#ff7f0e')
ax2.fill_between(df['wall_clock_time'],
                  rss_normalized - rss_normalized_std,
                  rss_normalized + rss_normalized_std,
                  alpha=0.2, color='#ff7f0e', label=f'RSS ±1 Std Dev (σ={rss_normalized_std:.2f})')

ax.set_xlabel('Wall Clock Time', fontsize=12)
ax.set_ylabel('CPU Percentage (%)', fontsize=12, color='#1f77b4')
ax2.set_ylabel('RSS (Normalized to 0-100)', fontsize=12, color='#ff7f0e')
ax.set_title('CPU Percentage vs RSS with Standard Deviation Bands', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.tick_params(axis='y', labelcolor='#1f77b4')
ax2.tick_params(axis='y', labelcolor='#ff7f0e')

# Combine legends
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```
