## Shapiro–Wilk normality test for Query A and Query B

The Shapiro–Wilk test is used to check whether the performance data for each query follows a normal (Gaussian) distribution. This is important because it determines which statistical tests are appropriate in later analysis.

The test is run separately for Query A and Query B.
A **p-value > 0.05** means the data does **not significantly deviate from normality**, while a **p-value ≤ 0.05** indicates that the data is **not normally distributed**.

Interpreting the results:
- If **both** Query A and Query B have p > 0.05 → the data is approximately normal, and parametric tests (like the t-test) can be used.
- If **either** query has p ≤ 0.05 → the data is non-normal, and non-parametric tests (such as Mann–Whitney U) should be used instead.

The Shapiro–Wilk test therefore serves as the first step in selecting the correct statistical method for comparing the performance of the two queries.

```python
shapiro_wilk(dict_a=traditional_dfs, dict_b=cn_dfs, parameter="elapsed_time")
shapiro_wilk(dict_a=traditional_dfs, dict_b=cn_dfs, parameter="process_cpu_percentage")
shapiro_wilk(dict_a=traditional_dfs, dict_b=cn_dfs, parameter="rss")
```

## Mann–Whitney U and Cliff’s Delta tests for comparing Query A and Query B

The Mann–Whitney U test is a non-parametric test used when the data is not normally distributed. It evaluates whether Query A and Query B come from the same distribution for a given performance metric (e.g., runtime, CPU usage, RAM). A p-value ≤ 0.05 indicates a statistically significant difference between the queries.

Cliff’s delta measures the **direction** and **magnitude** of that difference. It ranges from −1 to +1:

- Values near **0** → negligible difference.
- Values near **+1** → Query A’s values are almost always **higher** than Query B’s.
- Values near **−1** → Query A’s values are almost always **lower** than Query B’s.

Interpretation depends on the metric:
- **Runtime**: delta near −1 → Query A is consistently faster; delta near +1 → Query A is slower.
- **CPU/RAM**: delta near −1 → Query A uses less resources; delta near +1 → Query A uses more.

Together, the U test tells you *whether* the queries differ, and Cliff’s delta tells you *how much* and *in which direction*.

```python
mann_whitney_u_test(dict_a=traditional_dfs, dict_b=cn_dfs, parameter="elapsed_time")
mann_whitney_u_test(dict_a=traditional_dfs, dict_b=cn_dfs, parameter="process_cpu_percentage")
mann_whitney_u_test(dict_a=traditional_dfs, dict_b=cn_dfs, parameter="rss")
```
